```python
# 연습문제
# 남자 승객의 나이의 결측치는, 남자 승객의 나이의 평균으로 채웁니다.
# 여자 승객의 나이의 결측치는, 여자 승객의 나이의 평균으로 채웁니다.
# 결측치를 모두 채운 후 age 컬럼의 평균을 출력합니다.(검증코드 실행하여 올바른 값이 출력 되는지 확인)

# 남자 승객
condMale = (df1['sex'] == 'male')
# 남자 승객의 평균 나이
meanMaleAge =  df1.loc[condMale, 'age'].mean()
# 남자 승객의 나이 결측치 채우기
df1.loc[condMale, 'age'] = df1.loc[condMale, 'age'].fillna(meanMaleAge)

# 여자 승객
condFemale = (df1['sex'] == 'female')
# 여자 승객의 평균 나이
meanFemaleAge = df1.loc[condFemale, 'age'].mean()
# 여자 승객의 결측치 채우기
df1.loc[condFemale, 'age'] = df1.loc[condFemale, 'age'].fillna(meanFemaleAge)
```

```python
# 검증
assert df1['age'].isnull().sum() == 0
assert df1['age'].mean().round(5) == 29.73603
```
